### Make Python read my origional cleaned B1/B2 Visa data file

```python
df = pd.read_csv("COMM 217 Visa Data Cleaned.csv")
```

### Melting all countires into a long and skinny form

```python
df_melt = df.melt(
    id_vars="Country", var_name="Year", value_name="RefusalRate"
)
```

```python
df_melt["Year"] = pd.to_numeric(df_melt["Year"])
```

```python
df_melt = df_melt.sort_values(["Country", "Year"])
```

### Check and see if all countires are in a melt form by looking at the year row

```python
print(df_melt.head())
```

### Pick out China, India, Japan, and South Koera from the dataset for visualization prep

```python
target_countries = [
    "CHINA",
    "INDIA",
    "JAPAN",
    "KOREA, SOUTH"
]
```

```python
df_selected = df_melt[df_melt["Country"].isin(target_countries)]
```

### Export New shorted data into csv format for tableau visualization

```python
df_selected.to_csv("asia_selected.csv", index=False)
```
