```python
import pandas as pd
```

```python
df = pd.read_csv("COMM 217 Visa Data Cleaned.csv")
```

```python
df_merge = df.melt(
    id_vars="Country", var_name="Year", value_name="RefusalRate"
)
```

```python
df_merge["Year"] = pd.to_numeric(df_merge["Year"])
```

```python
df_merge = df_merge.sort_values(["Country", "Year"])
```

```python
print(df_merge.head())
```

```python
target_countries = [
    "CHINA",
    "INDIA",
    "JAPAN",
    "KOREA, SOUTH"
]
```

```python
df_selected = df_merge[df_merge["Country"].isin(target_countries)]
```

```python
df_selected.to_csv("asia_selected_merge.csv", index=False)
```
